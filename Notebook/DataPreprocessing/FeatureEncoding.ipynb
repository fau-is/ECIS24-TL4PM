# Feature Encoding

```python
import pandas as pd
import numpy as np
import os
import sys
module_path = os.path.abspath(os.path.join('../..'))
if module_path not in sys.path:
    sys.path.append(module_path)
import pickle as pickle

from sklearn.preprocessing import MinMaxScaler
```

## Read trace data

```python
split_pattern = ["811split", "641620split"]
data_set_type = ["source", "target"]
data_set_cat = ["train", "val", "test"]
split_strat = ["", "_alt"]
```

```python
folder_path = "../../Data/Training/"
for s_pattern in split_pattern:
    path = os.path.join(folder_path, s_pattern)
    for d_type in data_set_type:
        lapse_scaler = MinMaxScaler()
        interval_scaler = MinMaxScaler()
        for d_cat in data_set_cat:
            file_name = d_type + "_" + d_cat + ".pkl"
            file_path = os.path.join(path, file_name)
            print(file_path)
            trace = pd.read_pickle(file_path)
            print("num case delayed: ", trace[trace["Label"] == 0].shape[0])
            print("num case in-time: ", trace[trace["Label"] == 1].shape[0])
            
            if d_cat == "train":
                lapse_scaler.fit(trace["CaseLapse"].explode("CaseLapse").values.reshape((-1,1)))
                interval_scaler.fit(trace["CaseInterval"].explode("CaseInterval").values.reshape((-1,1)))
                     
            trace["CaseLapse"] = trace["CaseLapse"].apply(lambda x: np.squeeze(lapse_scaler.transform(x.reshape(-1, 1))))
            trace["CaseInterval"] = trace["CaseInterval"].apply(lambda x: np.squeeze(interval_scaler.transform(x.reshape(-1, 1))))
            trace["OverallCriteria"] = trace["OverallCriteria"].apply(lambda x: np.squeeze(lapse_scaler.transform(np.array([[x]]))))
            
            trace.to_pickle(file_path)
    
```

```python
folder_path = "../../training_data/"
for s_pattern in split_pattern:
    path = os.path.join(folder_path, s_pattern)
    for d_type in data_set_type:
        
        lapse_scaler = MinMaxScaler()
        interval_scaler = MinMaxScaler()
        
        trace_train = pd.read_pickle(os.path.join(path, d_type + "_train_alt.pkl"))
        print("num case delayed in train: ", trace_train[trace_train["Label"] == 0].shape[0])
        print("num case in-time in train: ", trace_train[trace_train["Label"] == 1].shape[0])
        
        trace_val = pd.read_pickle(os.path.join(path, d_type + "_val_alt.pkl"))
        print("num case delayed in val: ", trace_val[trace_val["Label"] == 0].shape[0])
        print("num case in-time in val: ", trace_val[trace_val["Label"] == 1].shape[0])
        
        trace_test = pd.read_pickle(os.path.join(path, d_type + "_test_alt.pkl"))
        print("num case delayed in test: ", trace_test[trace_test["Label"] == 0].shape[0])
        print("num case in-time in test: ", trace_test[trace_test["Label"] == 1].shape[0])
        
        trace_seen = pd.concat([trace_train, trace_val])
        
        lapse_scaler.fit(trace_seen["CaseLapse"].explode("CaseLapse").values.reshape((-1,1)))
        interval_scaler.fit(trace_seen["CaseInterval"].explode("CaseInterval").values.reshape((-1,1)))
        
        for trace in [trace_train, trace_val, trace_test]:
            trace["CaseLapse"] = trace["CaseLapse"].apply(lambda x: np.squeeze(lapse_scaler.transform(x.reshape(-1, 1))))
            trace["CaseInterval"] = trace["CaseInterval"].apply(lambda x: np.squeeze(interval_scaler.transform(x.reshape(-1, 1))))
            trace["OverallCriteria"] = trace["OverallCriteria"].apply(lambda x: np.squeeze(lapse_scaler.transform(np.array([[x]]))))

        trace_train.to_pickle(os.path.join(path, d_type + "_train_alt.pkl"))
        trace_val.to_pickle(os.path.join(path, d_type + "_val_alt.pkl"))
        trace_test.to_pickle(os.path.join(path, d_type + "_test_alt.pkl"))
```
