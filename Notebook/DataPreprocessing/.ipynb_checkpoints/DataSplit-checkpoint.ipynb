# Data Split

```python
import pandas as pd
import numpy as np
import os
import sys
module_path = os.path.abspath(os.path.join('../..'))
if module_path not in sys.path:
    sys.path.append(module_path)
import pickle as pickle
```

```python
source_trace = pd.read_pickle("../../Data/Trace/source_trace_all_l.pkl")
target_trace = pd.read_pickle("../../Data/Trace/target_trace_all_l.pkl")
```

## Split

```python
def split_data(data, ratio=[0.8, 0.1, 0.1], sort=True):
    data_size = data.shape[0]
    train_size = int(ratio[0] * data_size)
    val_size = int(ratio[1] * data_size)
    test_size = data_size - train_size - val_size
    if sort:
        data_sorted = data.sort_values(by='Start Time', ascending=True)
    else:
        data_sorted = data.sample(frac=1)
    data_train = data_sorted[:train_size]
    data_val = data_sorted[train_size: train_size+val_size]
    data_test = data_sorted[train_size+val_size:]
    return data_train, data_val, data_test
```

```python
source_train, source_val, source_test = split_data(source_trace, sort=False)
target_train, target_val, target_test = split_data(target_trace, sort=False)

source_train.to_pickle("../../Data/Training/811split/source_train_random_all.pkl")
source_val.to_pickle("../../Data/Training/811split/source_val_random_all.pkl")
source_test.to_pickle("../../Data/Training/811split/source_test_random_all.pkl")

target_train.to_pickle("../../Data/Training/811split/target_train_random_all.pkl")
target_val.to_pickle("../../Data/Training/811split/target_val_random_all.pkl")
target_test.to_pickle("../../Data/Training/811split/target_test_random_all.pkl")
```

```python
source_train, source_val, source_test = split_data(source_trace, ratio=[0.64,0.16,0.2], sort=False)
target_train, target_val, target_test = split_data(target_trace, ratio=[0.64,0.16,0.2], sort=False)

source_train.to_pickle("../../Data/Training/641620split/source_train_random_all.pkl")
source_val.to_pickle("../../Data/Training/641620split/source_val_random_all.pkl")
source_test.to_pickle("../../Data/Training/641620split/source_test_random_all.pkl")

target_train.to_pickle("../../Data/Training/641620split/target_train_random_all.pkl")
target_val.to_pickle("../../Data/Training/641620split/target_val_random_all.pkl")
target_test.to_pickle("../../Data/Training/641620split/target_test_random_all.pkl")
```

```python
def split_data_alt(data, ratio=[0.8, 0.1, 0.1], sort=True):
    data_size = data.shape[0]
    train_size = int(ratio[0] * data_size)
    test_size = int(ratio[2] * data_size)
    if sort:
        data_sorted = data.sort_values(by='Start Time', ascending=True)
    else:
        data_sorted = data.sample(frac=1)
    data_test = data_sorted[-test_size:]
    
    data_rest = data_sorted[:-test_size]
    data_rest = data_rest.sample(frac=1)
    data_train = data_rest[:train_size]
    data_val = data_rest[train_size:]
    return data_train, data_val, data_test
```

```python
source_train, source_val, source_test = split_data_alt(source_trace)
target_train, target_val, target_test = split_data_alt(target_trace)

source_train.to_pickle("../../Data/Training/811split/source_train_alt.pkl")
source_val.to_pickle("../../Data/Training/811split/source_val_alt.pkl")
source_test.to_pickle("../../Data/Training/811split/source_test_alt.pkl")

target_train.to_pickle("../../Data/Training/811split/target_train_alt.pkl")
target_val.to_pickle("../../Data/Training/811split/target_val_alt.pkl")
target_test.to_pickle("../../Data/Training/811split/target_test_alt.pkl")
```

```python
source_train, source_val, source_test = split_data_alt(source_trace, ratio=[0.64,0.16,0.2])
target_train, target_val, target_test = split_data_alt(target_trace, ratio=[0.64,0.16,0.2])

source_train.to_pickle("../../Data/Training/641620split/source_train_alt.pkl")
source_val.to_pickle("../../Data/Training/641620split/source_val_alt.pkl")
source_test.to_pickle("../../Data/Training/641620split/source_test_alt.pkl")

target_train.to_pickle("../../Data/Training/641620split/target_train_alt.pkl")
target_val.to_pickle("../../Data/Training/641620split/target_val_alt.pkl")
target_test.to_pickle("../../Data/Training/641620split/target_test_alt.pkl")
```
